```python
## Is there a correlation between the amount of daily mobile phone usage and performance impact?

## CHI-SQUARE TEST IS USED TO TEST THE RELATIONSHIP BETWEEN 2 CATEGORICAL VARAIBLES
from scipy.stats import chi2_contingency

#creation of contingency table
contingency_table = pd.crosstab(df['Daily usages'], df['Performance impact'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu")
plt.title('Contingency Table: Daily Usage vs Performance Impact')
plt.ylabel('Daily Usage')
plt.xlabel('Performance Impact')
plt.show()
```

```python
## How does the frequency of symptoms relate to the health rating given by participants?

contingency_table = pd.crosstab(df['Symptom frequency'], df['Health rating'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Symptom Frequency vs Health Rating')
plt.ylabel('Symptom Frequency')
plt.xlabel('Health Rating')
plt.show()
```

```python
## Are there any trends in mobile phone usage patterns based on the operating system?
contingency_table = pd.crosstab(df['Daily usages'], df['Mobile Operating System'])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt="d")
```
